```python
def fill_key_and_mode(row, df_non_nan, mode_value):
    # Filter rows by non-NaN mode and the specific mode (major or minor)
    relevant_rows = df_non_nan[(df_non_nan['mode'] == mode_value) & (df_non_nan['key'].notna())]
    
    # Find the row with the closest mean_score
    closest_row = relevant_rows.iloc[(relevant_rows['mean_score'] - row['mean_score']).abs().argsort()[:1]]
    
    # Fill the NaN values for key and mode based on the closest row
    row['key'] = closest_row['key'].values[0]
    return row
```

```python
#relacing the nan value for the key with the approximation of nearest key by comparing all the metrics and mode

df['mean_score'] = df[['danceability', 'energy', 'valence','acousticness','instrumentalness','liveness','speechiness']].mean(axis=1)

df_nan = df[df['key'].isna()]
df_non_nan = df[df['key'].notna()]

for idx, row in df_nan.iterrows():
    # Here we assume, for example purposes, that NaN row should be treated as major or minor
    # You could adjust this condition based on your data or another strategy
    if row['mode']=='Major':  # Assuming higher mean_score tracks are major
        df_nan.loc[idx] = fill_key_and_mode(row, df_non_nan, mode_value='Major')
    else:  # Assuming lower mean_score tracks are minor
        df_nan.loc[idx] = fill_key_and_mode(row, df_non_nan, mode_value='Minor')

df1 = pd.concat([df_non_nan, df_nan]).sort_index()
```

```python
#filling the nan values with 0 for in_shazam playlists
df1['in_shazam']=df1['in_shazam'].fillna(0)
df1['in_deezer_playlists']=df1['in_deezer_playlists'].fillna(0)
df1
```

```python
#top artists 
df1['artist_name']=df['artist_name'].str.split(",")
df2=df1.explode('artist_name')
total_streams=df2.groupby('artist_name')['streams'].sum().reset_index()
top_artists=total_streams.sort_values('streams',ascending=False).head(10)
top_artists
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.bar(top_artists['artist_name'], top_artists['streams'], color='skyblue')
plt.xlabel('Artists')
plt.ylabel('Total Number of Streams')
plt.title('Total Streams by Artist')
plt.xticks(rotation=45)
plt.tight_layout()  # Adjust layout for better spacing
plt.show()
```

```python
#how this top 10 artists from each year
df2['Year'] = df2['released_date'].dt.year

se=df2.groupby(['Year','artist_name'])['streams'].sum().reset_index()
top_artist_yearly=se.sort_values(['Year','streams'],ascending=[False,False])
res=top_artist_yearly.groupby('Year').head(10).reset_index(drop=True)
res
```

```python
#top 10 tracks each year
top_10_tracks=df2.groupby(['Year','track_name'])['streams'].sum().reset_index()
top_track_yearly=top_10_tracks.sort_values(['Year','streams'],ascending=[False,False])
top_10_per_year = top_track_yearly.groupby('Year').head(10).reset_index(drop=True)
top_10_per_year
```
